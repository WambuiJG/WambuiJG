The goal of this analysis is to identify key highlights and performance patterns from soccer games.The dataset covers every match from the 2018–2019 English Premier League season, providing detailed information on half‑time and full‑time outcomes, frequency of fouls, red cards and yellow cards.


```python
import pandas as pd
soccer = pd.read_csv("soccer18-19.csv")
print(soccer.shape)
soccer.head(100)
```

```python
# statistical summary
soccer.describe()
```

```python
#checking the datatypes
soccer.dtypes
```

```python
#checking for null values
soccer.isnull().sum()
```

```python
#finding most fouls given to a hometeam
most_fouls_HF=soccer[soccer['HF'] == soccer['HF'].max()]
most_fouls_HF.head()
```

```python
#finding most fouls given to an awayteam
most_fouls_AF=soccer[soccer['AF'] == soccer['AF'].max()]
most_fouls_AF.head()
```

```python
#Finding the Away team which commits the most fouls
AwayTeam_most_fouls=soccer[soccer['AF'] == soccer['AF'].max()]['AwayTeam']
print(AwayTeam_most_fouls)

number_of_the_most_fouls_AF=soccer[soccer['AF'] == soccer['AF'].max()]['AF']
print(number_of_the_most_fouls_AF)
```

```python
# Finding the Home team which commits the most fouls
HomeTeam_most_fouls=soccer[soccer['HF'] == soccer['HF'].max()]['HomeTeam']
print(HomeTeam_most_fouls)

number_of_the_most_fouls_HF=soccer[soccer['HF'] == soccer['HF'].max()]['HF']
print(number_of_the_most_fouls_HF)
```

```python
#percentage of games that ended in a draw over time
instances_with_draws=soccer[soccer['FTR'] == 'D']
percentage_draws=len(instances_with_draws)/len(soccer)*100
print(percentage_draws)
```

```python
#percentage of games that ended in a win for the home team over time
instances_with_home_wins=soccer[soccer['FTR'] == 'H']
percentage_home_wins=len(instances_with_home_wins)/len(soccer)*100
print(percentage_home_wins)
```

```python
#percentage of games that ended in a win for the away team over time
instances_with_away_wins=soccer[soccer['FTR'] == 'A']
percentage_away_wins=len(instances_with_away_wins)/len(soccer)*100
print(percentage_away_wins)
```

```python
#creating a pie chart
games_wins_distribution=[]
games_wins_distribution.append(percentage_draws)
games_wins_distribution.append(percentage_home_wins)
games_wins_distribution.append(percentage_away_wins)

import matplotlib.pyplot as plt
import numpy as np

mylabels = ["Draws", "Home Wins", "Away Wins"]
plt.pie(games_wins_distribution, labels = mylabels)
plt.show()
```

```python
#information on the half time draws
half_time_draws=soccer[soccer['HTR']=='D']

# Half time  Draws that become Full time draws
HTR_FTR_draws=half_time_draws[half_time_draws['FTR']=='D']

#Percentage of htr which is a draw that results in ftr which is a draw
percentage_HTR_FTR_draws=len(HTR_FTR_draws)/len(half_time_draws)*100
print(f"The percentage that a half-time draw result becomes a full-time draw is {percentage_HTR_FTR_draws}%. ")

#Half time draw become a win to home team
HTR_draws_FTR_home_wins=half_time_draws[half_time_draws['FTR']=='H']
percentage_HTR_draws_FTR_home_wins=len(HTR_draws_FTR_home_wins)/len(half_time_draws)*100
print(f"The percentage that a half-time draw result becomes a full-time win for the home team is {percentage_HTR_draws_FTR_home_wins}%. ")

#Half time draw become a win to away team
HTR_draws_FTR_away_wins=half_time_draws[half_time_draws['FTR']=='A']
percentage_HTR_draws_FTR_away_wins=len(HTR_draws_FTR_away_wins)/len(half_time_draws)*100
print(f"The percentage that a half-time draw result becomes a full-time win for the away team is {percentage_HTR_draws_FTR_away_wins}%. ")
```

```python
#half time wins to home team
half_time_home_wins=soccer[soccer['HTR']=='H']

# Half time  Wins to home team that become Full time wins for home team
HTR_FTR_home_wins=half_time_home_wins[half_time_home_wins['FTR']=='H']

percentage_HTR_FTR_home_wins=len(HTR_FTR_home_wins)/len(half_time_home_wins)*100
print(f"The percentage that a half-time lead for the home team is maintained until full-time is {percentage_HTR_FTR_home_wins}%. ")

#  Half time Wins to home team that become Full time draw
HTR_home_wins_FTR_draw=half_time_home_wins[half_time_home_wins['FTR']=='D']

percentage_HTR_home_wins_FTR_draw=len(HTR_home_wins_FTR_draw)/len(half_time_home_wins)*100
print(f"The percentage that a half-time lead for the home team becomes a full-time draw is {percentage_HTR_home_wins_FTR_draw}%. ")

#  Half time Wins to home team that become win to away team
HTR_home_wins_FTR_away_win=half_time_home_wins[half_time_home_wins['FTR']=='A']

percentage_HTR_home_wins_FTR_away_win=len(HTR_home_wins_FTR_away_win)/len(half_time_home_wins)*100
print(f"The percentage that a half-time lead for the home team becomes a full-time win for the away team is {percentage_HTR_home_wins_FTR_away_win}%. ")
```

```python
#half time win to away team that becomes full time win to away team
half_time_away_wins=soccer[soccer['HTR']=='A']

# Half time  Wins to away team that become Full time wins for away team
HTR_FTR_away_wins=half_time_away_wins[half_time_away_wins['FTR']=='A']

percentage_HTR_FTR_away_wins=len(HTR_FTR_away_wins)/len(half_time_away_wins)*100
print(f"The percentage that a half-time lead for the away team is maintained until full-time is {percentage_HTR_FTR_away_wins}%. ")

#  Half time Wins to away team that become Full time draw
HTR_away_wins_FTR_draw=half_time_away_wins[half_time_away_wins['FTR']=='D']

percentage_HTR_away_wins_FTR_draw=len(HTR_away_wins_FTR_draw)/len(half_time_away_wins)*100
print(f"The percentage that a half-time lead for the away team becomes a full-time draw is {percentage_HTR_away_wins_FTR_draw}%. ")

#  Half time Wins to away team that become win to home team
HTR_away_wins_FTR_home_win=half_time_away_wins[half_time_away_wins['FTR']=='H']

percentage_HTR_away_wins_FTR_home_win=len(HTR_away_wins_FTR_home_win)/len(half_time_away_wins)*100
print(f"The percentage that a half-time lead for the away team becomes a full-time win for the home team is {percentage_HTR_away_wins_FTR_home_win}%. ")
```
